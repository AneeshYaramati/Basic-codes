```python
import numpy as np
class vectors:
    def __init__(self,i,j,k):
        self.dir_i=i
        self.dir_j=j
        self.dir_k=k
        
    def add(self,v1,v2):
        self.add_i = v1.dir_i + v2.dir_i
        self.add_j = v1.dir_j + v2.dir_j
        self.add_k = v1.dir_k + v2.dir_k
        print("Resultant Vector after adding is: {:.2f}i + {:.2f}j+ {:.2f}k ".format(self.add_i,self.add_j,self.add_k))
    
    def dotProduct(self,v1,v2):
        self.listVector1 = [v1.dir_i,v1.dir_j,v1.dir_k]
        self.listVector2 = [v2.dir_i,v2.dir_j,v2.dir_k]
        self.dotProd = np.dot(self.listVector1,self.listVector2)
        print("DotProduct of two vectors is : {}".format(self.dotProd))
        
    def display(self):
        print("The vector input is: {}i + {}j + {}k ".format(self.dir_i,self.dir_j,self.dir_k))

print("Enter the first vector: ")
i = float(input("Enter x-corodinate of Vector1:"))
j = float(input("Enter y-corodinate of Vector1:"))
k = float(input("Enter z-corodinate of Vector1:"))
        
v1 = vectors(i,j,k)
v1.display()
print("Enter the Second vector: ")
i2 = float(input("Enter x-corodinate of Vector2:"))
j2 = float(input("Enter y-corodinate of Vector2:"))
k2 = float(input("Enter z-corodinate of Vector2:"))
v2 = vectors(i2,j2,k2)
v2.display()

v3= vectors(0,0,0)
v3.add(v1,v2)
v3.dotProduct(v1,v2)

    
```

```python
class internalMarks:
    def __init__(self,p1,p2,e1,e2,e3):
        self.p1 = p1
        self.p2 = p2
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3
    
    def displayInternal(self):
        print("Periodical-1: {:.2f}".format(self.p1))
        print("Periodical-2: {:.2f}".format(self.p2))
        print("Lab Evaluation-1: {:.2f}".format(self.e1))
        print("Lab Evaluation-2: {:.2f}".format(self.e2))
        print("Lab Evaluation-3: {:.2f}".format(self.e3))
    
class externalMarks(internalMarks):
    def __init__(self,on_exam,on_viva):
        self.on_exam = on_exam
        self.on_viva = on_viva
    
    def displayExternal(self):
        print("Online Exam: {:.2f}".format(self.on_exam))
        print("Online Viva: {:.2f}".format(self.on_viva))
    
class Student(externalMarks):
    def __init__(self,name,roll):
        self.name = name
        self.roll = roll
    
    def displayStudent(self):
        print("Student Name: "+self.name)
        print("Roll Number: "+self.roll)


class grade(Student):
    def __init__(self,name,roll,on_exam,on_viva,p1,p2,e1,e2,e3):
        Student.__init__(self,name,roll)
        externalMarks.__init__(self,on_exam,on_viva)
        internalMarks.__init__(self,p1,p2,e1,e2,e3)
    
    def gradeCalculation(self):
        self.total = sum([self.on_exam,self.on_viva,self.p1,self.p2,self.e1,self.e2,self.e3])
        print("Total Marks: {:.2f}".format(self.total))
        if self.total>=91 and self.total<=100:
            print("grade: A")
        
        elif self.total>=81 and self.total<=90:
            print("grade: B")
        
        elif self.total>=61 and self.total<=80:
            print("grade: P")
        
        else:
            print("grade: F")
    
```

```python
name = str(input("Enter Name:"))
roll = str(input("Enter Roll-No:"))
on_exam = float(input("Enter online Exam mark:"))
on_viva = float(input("Enter Online Viva mark:"))
p1 = float(input("Enter P1-mark:"))
p2 = float(input("Enter P2-mark:"))
e1= float(input("Enter Evaluation-1 mark:"))
e2= float(input("Enter Evaluation-2 mark:"))
e3 = float(input("Enter Evaluation-3 mark:"))

a = grade(name,roll,on_exam,on_viva,p1,p2,e1,e2,e3)
print("\n")
a.displayStudent()
a.displayExternal()
a.displayInternal()
a.gradeCalculation()
```
